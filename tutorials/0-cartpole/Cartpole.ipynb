```python
import gym
```

```python
env = gym.make('CartPole-v0')
```

```python
for i_episode in range(10):
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()
        print(action)
        observation, reward, done, info = env.step(t%2)

env.close()

```
